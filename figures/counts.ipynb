# Calculate the number of choices and plans that both subjects made

```python
import matplotlib.pyplot as plt
import utils, plotting, graph
import numpy as np
import seaborn as sns
from dataClass import nwbWrapper

def bhv_counts():
    filenames = utils.get_filenames()
    
    def get_counts(fname):
        data = nwbWrapper(fname, region="OFC", to_load="bhv")
        n_choices = data.choice_df.query("trialerror == 0").shape[0]
        n_plans = data.fixation_df.query("(trialerror == 0) & (active_prob <= 0.2) & (duration >= 50) & (step_on <= 100)").shape[0]
        return dict(n_choices=n_choices, n_plans=n_plans, )

    res = utils.iterate_subjects(filenames, get_counts)
    return res

res = bhv_counts()
import pandas as pd
for key in res:
    res[key] = pd.DataFrame(res[key])
```

```python
for key in res:
    out = res[key].agg(["sum", "mean", "sem"])
    for stat in out.index:
        print(f"{key} {stat}:")
        print("Choices: %d, Plans: %d" % (out.loc[stat].values[0], out.loc[stat].values[1]))
        print()
```
